### Image Processing Playground

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

upload = widgets.FileUpload(accept='image/*', multiple=False)

grayscale = widgets.Checkbox(description="Сделать grayscale")
blur = widgets.Checkbox(description="Сделать blur")
edge = widgets.Checkbox(description="Сделать edge detection")
resize = widgets.Checkbox(description="Сделать resize")
rotation = widgets.Checkbox(description="Сделать rotation")

blur_kernel = widgets.IntSlider(value=5, min=1, max=31, step=2, description="Blur kernel")

edge_min = widgets.IntText(value=100, description="Canny min")
edge_max = widgets.IntText(value=200, description="Canny max")

resize_w = widgets.IntText(value=300, description="Width")
resize_h = widgets.IntText(value=300, description="Height")

angle = widgets.IntText(value=90, description="Angle")

button = widgets.Button(description="Run")
output = widgets.Output()

display(
    upload,
    grayscale,
    blur, blur_kernel,
    edge, edge_min, edge_max,
    resize, resize_w, resize_h,
    rotation, angle,
    button,
    output
)

def show_image(image):
    plt.figure(figsize=(6, 6))

    if len(image.shape) == 2:
        plt.imshow(image, cmap="gray")
    else:
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        plt.imshow(image_rgb)

    plt.axis("off")
    plt.show()

def rotate_any_angle(image, angle_value):
    h, w = image.shape[:2]
    center = (w // 2, h // 2)

    matrix = cv2.getRotationMatrix2D(center, angle_value, 1.0)
    rotated = cv2.warpAffine(image, matrix, (w, h))

    return rotated

def on_click(b):
    with output:
        clear_output()

        if not upload.value:
            print("Сначала загрузи изображение")
            return

        uploaded_file = upload.value[0]
        content = uploaded_file["content"]

        image_array = np.frombuffer(content, np.uint8)
        img = cv2.imdecode(image_array, cv2.IMREAD_COLOR)

        result = img.copy()

        if grayscale.value:
            result = cv2.cvtColor(result, cv2.COLOR_BGR2GRAY)

        if blur.value:
            result = cv2.GaussianBlur(result, (blur_kernel.value, blur_kernel.value), 0)

        if edge.value:
            if len(result.shape) == 3:
                gray = cv2.cvtColor(result, cv2.COLOR_BGR2GRAY)
            else:
                gray = result

            result = cv2.Canny(gray, edge_min.value, edge_max.value)

        if resize.value:
            result = cv2.resize(result, (resize_w.value, resize_h.value))

        if rotation.value:
            result = rotate_any_angle(result, angle.value)

        show_image(result)

button.on_click(on_click)
```
